```python
f = open("input_q1.txt")
with open("input_q1.txt", 'r') as file:
        rotations = [line.strip() for line in file]
```

```python
rotations[0]
```

```python
last = 50
answers = []
for i in range(len(rotations)):
    if rotations[i][0] == 'R':
        last = np.mod(last + int(rotations[i][1:]),100)
        answers .append(last)
    else:
        last = np.mod(last - int(rotations[i][1:]), 100 )
        answers .append(last)
```

```python
## final answer
print('final answer:' , len([i for i in answers if i == 0]))
```
